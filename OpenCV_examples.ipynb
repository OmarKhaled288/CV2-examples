```python
import cv2 as cv
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
path = '/content/0008f3d3-2f85-4973-be9a-1b520b8b59fc___JR_HL 4092.JPG'
```

```python
image = cv.imread(path)
```

```python
cv2_imshow(image)
```

```python
h, w, c = image.shape
w = int(w * 2)
h = int(h * 2)
dim = (w, h)
resized = cv.resize(image, dim, interpolation = cv.INTER_AREA)
cv2_imshow(resized)
```

```python
resized[0:100, 0:100] = (255,255,0) 
cv2_imshow(resized)
```

```python
cv.line(resized, (0,0), (resized.shape[1], resized.shape[0]), (0, 0, 255), 10)
cv2_imshow(resized)
```

```python
cv.rectangle(resized, (0,0), (resized.shape[1], resized.shape[0]), (0, 0, 255), 10)
cv2_imshow(resized)
```

```python
cv.circle(resized, (int(resized.shape[1] / 2), int(resized.shape[0] / 2)), 75, (0, 0, 255), 10)
cv2_imshow(resized)
```

```python
translatioMatrix = np.float32([[1, 0, 50], [0, 1, 50]])
resized = cv.warpAffine(resized, translatioMatrix, (resized.shape[1], resized.shape[0]))
cv2_imshow(resized)
```

```python
rotationMatrix = cv.getRotationMatrix2D((int(resized.shape[1] / 2), int(resized.shape[0] / 2)), 30, 1)
resized = cv.warpAffine(resized, rotationMatrix, (resized.shape[1], resized.shape[0]))
cv2_imshow(resized)
```

```python
resized = cv.flip(resized, 1)
cv2_imshow(resized)
```

```python
resized = cv.flip(resized, 0)
cv2_imshow(resized)
```

```python
resized = resized[0:int(resized.shape[1] / 2), 0:int(resized.shape[0] / 2)]
cv2_imshow(resized)
```

```python
resized = cv.add(resized, image)
cv2_imshow(resized)
```

```python
resized = resized * 255
cv2_imshow(resized)
```

```python
resized = cv.addWeighted(resized, 0.7, image, 0.3, 0)
cv2_imshow(resized)
```

```python
resized = resized - image
cv2_imshow(resized)
```
